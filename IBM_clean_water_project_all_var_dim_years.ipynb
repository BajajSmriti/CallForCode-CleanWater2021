## PCA with countries as features

```python
dimensions = {}
count = 0
for i in final_clean_df_var_T_pro.T.index:
    dimensions[count] = i
    count += 1
```

```python
dimensions
```

```python
pca_countries_dim = PCA(n_components=200)
```

```python
pca_countries_dim.fit(final_clean_df_var_T_pro)
```

```python
pca_countries_dim_comps = pca_countries_dim.fit_transform(final_clean_df_var_T_pro)
```

```python
pca_countries_dim_comps.shape
```

```python
m = pca_countries_dim.explained_variance_ratio_

# plotting code
fig = plt.figure(figsize=(10,5))
ax = plt.gca()
ax.set_xlabel("K")
ax.set_ylabel("Eigen Value")
ax.plot(m, '-o')
```

```python
# write your code here
count = 1
K = 0
K_var = {}
total_var = 0
for i in m:
    total_var = total_var + i
    K_var[K] = total_var
    if total_var <= 0.50:
        count += 1
    K += 1

print("Number of components needed to explain around 50% of the total variance : ",(count))
K_var_list = []
for i in K_var.values():
    K_var_list.append(i)

# plotting code
fig = plt.figure(figsize=(10,5))
ax = fig.gca()
ax.set_xlabel("K")
ax.set_ylabel("Total Variance Explained")
ax.plot(K_var_list, '-o')
```

```python
pca_countries_dim.components_.shape
```

```python
print('The estimated number of components. :' , pca_countries_dim.n_components_)
print('Number of features in the training data :' , pca_countries_dim.n_features_)
print('Number of samples in the training data :' , pca_countries_dim.n_samples_)
```

```python
eigen_vectors = pca_countries_dim.components_.T
```

```python
# creating U of dxk dimensions with components having absolute larger than 0.20 of the norm
count = 0
U_dxk = []
arr = []
for i in range(200):
    col = []
    for j in range(len(eigen_vectors[0])):
        if abs(eigen_vectors[j][i]) > (0.2*sum(eigen_vectors[i]**2)):
            col.append(1)
        else:
            col.append(0)
    U_dxk.append(col)

#fetching respective words for those components
top_50_pc_dim = dict()
c = 1
for x in U_dxk:
    for i in range(200):
        if x[i] == 1:
            if c not in top_50_pc_dim:
                top_50_pc_dim[c] = [dimensions[i]]
            else:
                top_50_pc_dim[c].append(dimensions[i])        
    c += 1

print('For the top 10 principal components, below Variables are important:- ')
top_50_pc_dim
```
